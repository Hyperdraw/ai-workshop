```python
import numpy as np
import brain
import matplotlib.pyplot as plt
from sklearn.preprocessing import minmax_scale
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
X = np.load('X.npy')
Y = np.load('y.npy')
Y_original = np.load('y.npy')
print(X.shape, Y.shape)
```

```python
Y = to_categorical(Y)
print(Y.shape)
```

```python
X = minmax_scale(X)
```

```python
X_train, X_test, Y_train, Y_test, _, Y_original_test = train_test_split(X, Y, Y_original, train_size=32)
```

```python
widths = [16] * 8 + [10]
params = np.random.rand(16 * (400 + 1) + (16 * (16 + 1)) * 7 + 10 * (16 + 1)) * 1.9 - 1
brain.gradient_descent(X_train, Y_train, brain.dense_relu_net_softmax, widths, brain.crossentropy, params, iterations=20, learning_rate=0.001, differentiation_ratio=10 ** -10, print_period=1, batch_size=8, pool_size=16)
```

```python
predictions = brain.dense_relu_net_softmax(X_test, widths, params)
(np.argmax(predictions, axis=1) == Y_original_test).mean()
```
